```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import VotingClassifier

# ‚úÖ Load training data
train_df = pd.read_csv("Train_Data (1).csv")
train_df.drop(columns=['SEQN'], inplace=True)
train_df = train_df[train_df['PAQ605'] != 7]

# Encode categorical variables
train_df['age_group'] = train_df['age_group'].map({'Adult': 0, 'Senior': 1})
train_df['RIAGENDR'] = train_df['RIAGENDR'].map({1: 1, 2: 0})
train_df['PAQ605'] = train_df['PAQ605'].map({1: 1, 2: 0})
train_df['DIQ010'] = train_df['DIQ010'].map({1: 1, 2: 0, 3: 1})

# Fill missing values
for col in ['RIAGENDR', 'PAQ605', 'DIQ010']:
    train_df[col] = train_df[col].fillna(train_df[col].mode()[0])

num_cols = ['BMXBMI', 'LBXGLU', 'LBXGLT', 'LBXIN']
train_df[num_cols] = train_df[num_cols].fillna(train_df[num_cols].median())
train_df.dropna(inplace=True)

# Features and target
y = train_df['age_group']
X = train_df.drop(columns=['age_group'])

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# üß™ SMOTE to balance classes
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_scaled, y)

# ‚úÖ Logistic Regression (Grid Search)
log_params = {
    'C': [0.01, 0.1, 1],
    'penalty': ['l1', 'elasticnet'],
    'l1_ratio': [0.3, 0.5],
    'solver': ['saga'],
    'class_weight': ['balanced']
}
log_grid = GridSearchCV(LogisticRegression(max_iter=2000), log_params, cv=5, scoring='f1_macro')
log_grid.fit(X_res, y_res)
log_best = log_grid.best_estimator_

print("\n‚úÖ Logistic Regression Best Params:", log_grid.best_params_)
print("üìä Logistic Regression Best F1 Score:", log_grid.best_score_)

# ‚úÖ XGBoost Classifier
scale_pos = (y == 0).sum() / (y == 1).sum()
xgb_model = XGBClassifier(
    n_estimators=100,
    max_depth=4,
    learning_rate=0.1,
    use_label_encoder=False,
    eval_metric='logloss',
    scale_pos_weight=scale_pos,
    random_state=42
)
xgb_model.fit(X_res, y_res)

# ‚úÖ Voting Classifier
voting_model = VotingClassifier(
    estimators=[
        ('logreg', log_best),
        ('xgb', xgb_model)
    ],
    voting='soft'
)
voting_model.fit(X_res, y_res)

# ‚úÖ Load and process external test set
test_df = pd.read_csv("Test_Data (1).csv")
test_df.drop(columns=['SEQN', 'RIDAGEYR'], inplace=True, errors='ignore')
test_df['RIAGENDR'] = test_df['RIAGENDR'].map({1: 1, 2: 0})
test_df['PAQ605'] = test_df['PAQ605'].map({1: 1, 2: 0})
test_df['DIQ010'] = test_df['DIQ010'].map({1: 1, 2: 0, 3: 1})

for col in ['RIAGENDR', 'PAQ605', 'DIQ010']:
    test_df[col] = test_df[col].fillna(test_df[col].mode()[0])
test_df[num_cols] = test_df[num_cols].fillna(test_df[num_cols].median())
test_df.dropna(inplace=True)

X_final_test = scaler.transform(test_df)
final_preds = voting_model.predict(X_final_test)

# Save prediction
pd.DataFrame({"age_group": final_preds.astype(int)}).to_csv("submission11.csv", index=False)
print("\nüìÅ Final predictions saved to 'submission.csv'")
```
